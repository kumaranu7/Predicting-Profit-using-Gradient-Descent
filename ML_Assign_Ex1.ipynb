```python
data = np.loadtxt('ex1data1.txt', delimiter = ',')
```

```python
X = np.c_[np.ones(data.shape[0]),data[:,0]] #adding a column of 1
```

```python
y = np.c_[data[:, 1]]
```

```python
plt.scatter(X[:,1],y, color = 'r',s = 30, marker = 'x', linewidths = 1)
plt.xlim(4,24)
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
```

```python
def ComputeCost(X, y, theta = [[0],[0]]):
    m = y.size
    h = X.dot(theta)
    J = 0.00
    J = (1/(2*m))* np.sum(np.square(h -y))
    return J
```

```python
ComputeCost(X,y)
```

```python
#to verify that gradient descent is working properly.
```

```python
def gradientDescent(X, y, theta=[[0],[0]], alpha=0.01, num_iters=1500):
    m = y.size
    J_history = np.zeros(num_iters)
    
    for iter in np.arange(num_iters):
        h = X.dot(theta)
        theta += - alpha*(1/m)*(X.T.dot(h-y))
        J_history[iter] = computeCost(X, y, theta)
    return(theta, J_history)
```

```python
theta , Cost_J = GradientDescent(X, y)
print('theta: ',theta.ravel())

plt.plot(Cost_J)
plt.ylabel('Cost J')
plt.xlim(0,1600)
plt.xlabel('Iterations')
```

```python
xx = np.arange(5,25)
yy = theta[0] + xx*theta[1]

#plotting gradient descent
plt.scatter(X[:,1] , y,s = 30, c = 'r', linewidths = 1, marker = 'x')
plt.plot(xx, yy,c = 'g', label = 'LinearRegression(Gradient Descent)')

#comparing with linearregression scikit learn
from sklearn.linear_model import LinearRegression
Regr = LinearRegression()
Regr.fit(X[:,1].reshape(-1,1), y.ravel())
plt.plot(xx, Regr.intercept_+Regr.coef_*xx, label='Linear regression (Scikit-learn GLM)')
```
